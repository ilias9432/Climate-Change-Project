```python
temp5 = temp5.reset_index()
```

```python
temp6 = temp5[(temp5['Year Series'] < 2036)]
```

```python
temp6_min = temp6[(temp6['Year Series'] == 2020)] 
temp6_max = temp6[(temp6['Year Series'] == 2035)] 
```

```python
temp6_merge = temp6_min.merge(temp6_max,on='Country',how='left')
```

```python
temp6_merge.rename({'Temp_x': '2020 Temp', 'Temp_y': '2038 Temp'}, axis=1, inplace=True)
temp6_merge = temp6_merge[['Country','2020 Temp','2038 Temp']]
```

```python
temp6_merge['Heat Wave'] = temp6_merge['2038 Temp'] - temp6_merge['2020 Temp']
temp6_merge.rename({'2020 Temp': '2020 Heat Wave'}, axis=1, inplace=True)
temp6_merge.rename({'2038 Temp': '2038 Heat Wave'}, axis=1, inplace=True)
```
